#### Indexing

You can use .reshape() to change the shape of a 1d array to a 2d,3d, etc.. array

Keep in mind, we will mainly be working with 2d tabular data.  

```python
mat = np.arange(0,100).reshape(10,10) # reshape 1x100 to 10x10
print(mat)
```

```python
row = 0
col = 1

# Selecting an individual number:
print(mat[row,col])

# Select an entire column (all row entries of this column "col")
print(mat[:,col])

# Select an entire column (all row entries of this column "col")
print(mat[row,:])
```

#### Masking

Masking allows you to use conditional filters to grab elements we'll see this idea used in pandas.

```python
print(mat > 50) # print bools

print(mat[mat>50]) # print direct vals from mat[]
```

### 2.2 Scatter Plots

- A line chart displays a series of data points (markers) connected by line segments.
- It is similiar to a scatter plot except that the measurement points are ordered (typically by their x-axis value) and joined with straight line segments.
- Often used to visualize a trend in data over intervals of time - know as a time series

This plots 100 random data points (set the seed to 42 to obtain the same points we do!) between 1 and 100 in both vertical and horizontal directions.

```python
import plotly.offline as pyo
import plotly.graph_objs as go
import numpy as np

np.random.seed(42)
random_x = np.random.randint(1,101,100) # from 1-101 return 100 random
random_y = np.random.randint(1,101,100)

data = [go.Scatter( # graph is build as a data list
    x = random_x,
    y = random_y,
    mode = 'markers', # can leave trailing comma
)]

pyo.plot(data, filename='scatter1.html')

IPython.display.HTML(filename='scatter1.html')
```

This plots 100 random data points (set the seed to 42 to obtain the same points we do!) between 1 and 100 in both vertical and horizontal directions.

```python
layout = go.Layout(
    title = 'Random Data Scatterplot', # Graph title
    xaxis = dict(title = 'Some random x-values'), # x-axis label
    yaxis = dict(title = 'Some random y-values'), # y-axis label
    hovermode ='closest' # handles multiple points landing on the same vertical
)
fig = go.Figure(data=data, layout=layout) # add layout to data as a fig
pyo.plot(fig, filename='scatter2.html')

IPython.display.HTML(filename='scatter2.html')
```

This plots 100 random data points (set the seed to 42 to obtain the same points we do!) between 1 and 100 in both vertical and horizontal directions.

```python
data = [go.Scatter(
    x = random_x,
    y = random_y,
    mode = 'markers',
    marker = dict( # change the marker style
        size = 12, # for marker
        color = 'rgb(51,204,153)',
        symbol = 'pentagon', # shape
        line = dict(
            width = 2, # line width outside of marker/symbol
        )
    ) # end marker
)]
layout = go.Layout(
    title = 'Random Data Scatterplot', # Graph title
    xaxis = dict(title = 'Some random x-values'), # x-axis label or xaxis = {"title": "Some random x-values"}
    yaxis = dict(title = 'Some random y-values'), # y-axis label
    hovermode ='closest' # handles multiple points landing on the same vertical
)
fig = go.Figure(data=data, layout=layout)
pyo.plot(fig, filename='scatter3.html')

IPython.display.HTML(filename='scatter3.html')
```

### 2.3 Line Charts

The file line2.py uses a customized dataset that
reduces U.S. Census Bureau statistics down to just
six New England states. To see how pandas was used
to trim the original data, see line3.py.


This line chart displays the same data three different ways along the y-axis.

```python
import plotly.offline as pyo
import plotly.graph_objs as go
import numpy as np

np.random.seed(56)
x_values = np.linspace(0, 1, 100) # 100 evenly spaced values
y_values = np.random.randn(100)   # 100 random values

# create traces
trace0 = go.Scatter(
    x = x_values,
    y = y_values+5, # add 5 to the random y values
    mode = 'markers', # set shape
    name = 'markers' # label name
)
trace1 = go.Scatter(
    x = x_values,
    y = y_values,
    mode = 'lines+markers',
    name = 'lines+markers'
)
trace2 = go.Scatter(
    x = x_values,
    y = y_values-5,
    mode = 'lines',
    name = 'lines'
)

data = [trace0, trace1, trace2]  # assign traces to data
layout = go.Layout(title = 'Line chart showing three different modes')

fig = go.Figure(data=data,layout=layout)
pyo.plot(fig, filename='line1.html')

IPython.display.HTML(filename='line1.html')
```

This line chart shows U.S. Census Bureau population data from six New England states.

```python
 # read a .csv file into a pandas DataFrame:
df = pd.read_csv('https://people.cs.vt.edu/cluersen/Plotly%20Dashboards%20with%20Dash%20Master%20Data/population.csv',
                 index_col=0)

# create traces
traces = [go.Scatter(
    x = df.columns,
    y = df.loc[name],
    mode = 'markers+lines',
    name = name
) for name in df.index]

layout = go.Layout(
    title = 'Population Estimates of the Six New England States'
)

fig = go.Figure(data=traces,layout=layout)
pyo.plot(fig, filename='line2.html')

IPython.display.HTML(filename='line2.html')
```

This line chart shows U.S. Census Bureau population data from six New England states. 

> This plot uses Pandas to extract desired data from the source


```python
df = pd.read_csv('https://people.cs.vt.edu/cluersen/Plotly%20Dashboards%20with%20Dash%20Master%20SourceData/nst-est2017-alldata.csv')
# Alternatively:
# df = pd.read_csv('https://www2.census.gov/programs-surveys/popest/datasets/2010-2017/national/totals/nst-est2017-alldata.csv')

# grab just the six New England states:
df2 = df[df['DIVISION']=='1'] # filter out data frame for if the 'division' column equals 1 then assign to df2
# set the index to state name:
df2.set_index('NAME', inplace=True) # grab a column from df2 and reset the numbering based off it, inplace=True is required else it would be df2 = df2.set_index("NAME")
# grab just the population columns:
df2 = df2[[col for col in df2.columns if col.startswith('POP')]] # itterate through df2.columns[] for each col add it to list_of_pop_col if it starts with "POP"
print(df2)
# traces = [go.Scatter(x=,y=,mode=,name=) for name in df.index] # four things you typically need to fill out
traces=[go.Scatter(
    x = df2.columns, # make dependent variable
    y = df2.loc[name], # make independent variable (state name)
    mode = 'markers+lines',
    name = name
) for state in df2.index] # build a scatterplot for every state in df2.index

layout = go.Layout(
    title = 'Population Estimates of the Six New England States'
)

fig = go.Figure(data=traces,layout=layout)
pyo.plot(fig, filename='line3.html')

IPython.display.HTML(filename='line3.html')
```

### 3.1 Dash Layout

```python
# -*- coding: utf-8 -*-

import dash
from dash import dcc
import dash_html_components as html

app = dash.Dash()

app.layout = html.Div(children=[ # children list so everything goes inside div
    # add some HTML elements
    html.H1(children='Hello Dash'),
    html.Div(children='Dash: A web application framework for Python.'),

    dcc.Graph(
        id='example-graph', # html id for graph element
        figure={ # takes in the actuace dash ficture
            'data': [
                {'x': [1, 2, 3], 'y': [4, 1, 2], 'type': 'bar', 'name': 'SF'},
                {'x': [1, 2, 3], 'y': [2, 4, 5], 'type': 'bar', 'name': 'Montréal'},
            ],
            'layout': {
                'title': 'Dash Data Visualization'
            }
        }
    )
])

if __name__ == '__main__':
    app.run_server()
```
